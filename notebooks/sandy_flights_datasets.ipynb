```python
import numpy as np
import pandas as pd
```

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
csv_file = "resources/airport_data.csv"
all_data_df = pd.read_csv(csv_file)
all_data_df
```

## Passenger dataset

```python
passengers_df = all_data_df[["State Name", "2019 Passengers (10% sample)"]]
passengers_df = passengers_df.replace(',','', regex=True)
passengers_df["2019 Passengers (10% sample)"] = pd.to_numeric(passengers_df["2019 Passengers (10% sample)"])
passengers_df = passengers_df[0:434]
passengers_grouped_df = passengers_df.groupby("State Name").sum()
passengers_sorted_df = passengers_grouped_df.sort_values("2019 Passengers (10% sample)", ascending=False)
passengers_cleaned_df = passengers_sorted_df.reset_index()
passengers_cleaned_df = passengers_cleaned_df.rename(columns = {"State Name":"state",
                                                                "2019 Passengers (10% sample)":"passenger_amount"})
passengers_cleaned_df
```

```python
passengers_cleaned_df.to_csv("resources/total_passengers.csv")
```

## Airfare dataset

```python
airfare_df = all_data_df[["State Name", "Average Fare ($)"]]
airfare_df = airfare_df[0:434]
airfare_grouped_df = airfare_df.groupby("State Name").mean()
airfare_sorted_df = airfare_grouped_df.sort_values("Average Fare ($)", ascending=False)
airfare_cleaned_df = airfare_sorted_df.reset_index()
airfare_cleaned_df = airfare_cleaned_df.rename(columns = {"State Name":"state",
                                                          "Average Fare ($)":"average_airfare"})
airfare_cleaned_df
```

```python
# exports df to csv file
airfare_cleaned_df.to_csv("resources/avg_airfare.csv")
```
