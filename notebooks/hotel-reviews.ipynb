```python
import pandas as pd
```

```python
csv_file = "resources/Datafiniti_Hotel_Reviews_Jun19.csv"

hotelreviews_df = pd.read_csv(csv_file)

del hotelreviews_df['reviews.dateAdded']

hotelreviews_df = hotelreviews_df.dropna(how="any")

print(hotelreviews_df.count())

hotelreviews_df.head()
```

```python
state_counts = hotelreviews_df["province"].value_counts()

print(state_counts)
```

```python
grouped_hotelreviews_df = hotelreviews_df.groupby(['id', 'name', 'city', 'latitude', 'longitude', 'province'])

grouped_hotelreviews_df.count().head(10)
```

```python
hotel_avg_ratings = grouped_hotelreviews_df["reviews.rating"].mean()
hotel_count_ratings = grouped_hotelreviews_df["reviews.rating"].count()

hotelratings_summary_df = pd.DataFrame({"ratings_average": hotel_avg_ratings,
                                    "ratings_count": hotel_count_ratings})

hotelratings_summary_df
```

```python
hotelratings_summary_df.to_csv("Resources/hotelratings.csv")
```
