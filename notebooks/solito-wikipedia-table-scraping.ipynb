```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pymongo
from splinter import Browser
import pandas as pd
import time
```

```python
url1 = "https://en.wikipedia.org/wiki/List_of_casinos_in_the_United_States#State"

response1 = requests.get(url1)

soup = BeautifulSoup(response1.text, 'html.parser')

w_table = soup.find('table')
w_tbody = w_table.find('tbody')

w_trs = w_tbody.find_all("tr")

w_trs
```

```python
casinos = []
cities = []
counties = []
states = []
districts = []
types = []
comments = []

for x in range(1, len(w_trs)):
    casinos.append(w_trs[x].find_all("td")[0].text)
    cities.append(w_trs[x].find_all("td")[1].text)
    counties.append(w_trs[x].find_all("td")[2].text)
    try:
        states.append(w_trs[x].find_all("td")[3].text)
    except:
        states.append("Delaware")
    try:
        districts.append(w_trs[x].find_all("td")[4].text)
    except:
        districts.append("")
    try:
        types.append(w_trs[x].find_all("td")[5].text)
    except:
        types.append("")
    try:
        comments.append(w_trs[x].find_all("td")[6].text)
    except:
        comments.append("")
        

        
```

```python
url2 = "https://en.wikipedia.org/wiki/List_of_national_parks_of_the_United_States"

response2 = requests.get(url2)

soup = BeautifulSoup(response2.text, 'html.parser')

w_table = soup.find('table', class_="wikitable")
w_tbody = w_table.find('tbody')

w_trs = w_tbody.find_all("tr")

w_trs
```

```python
parks = []
locations = []
dates = []
areas = []
visitors = []

for x in range(1, len(w_trs)):
    try:
        parks.append(w_trs[x].find("th").find("a").text)
        locations.append(w_trs[x].find_all("td")[1].text.strip())
        dates.append(w_trs[x].find_all("td")[2].text.strip())
        areas.append(w_trs[x].find_all("td")[3].text.strip())
        visitors.append(w_trs[x].find_all("td")[4].text.strip())
    except:
        parks.append(w_trs[x].find_all("td")[0].find("a").text)
        locations.append(w_trs[x].find_all("td")[2].text.strip())
        dates.append(w_trs[x].find_all("td")[3].text.strip())
        areas.append(w_trs[x].find_all("td")[4].text.strip())
        visitors.append(w_trs[x].find_all("td")[5].text.strip())

parks_df = pd.DataFrame({"park":parks,"location":locations,"date_established":dates,"area":areas,"visitors":visitors})

parks_df
```
